```python
import torchvision.models as models
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
```

```python
def get_parameters_count(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)
```

```python
modeltotal=[
    models.resnet18(),
    models.resnet34(),
    models.resnet50(),
    models.resnet101(),
    models.resnet152(),
    models.vgg11(),
    models.vgg11_bn(),
    models.vgg13(),
    models.vgg13_bn(),
    models.vgg16(),
    models.vgg16_bn(),
    models.vgg19()
]
total_parameters=[]
model_name=[]
for model in modeltotal:
    total_parameters.append(get_parameters_count(model))
    model_name.append(type(model).__name__)
```
